```
# print the numbers from 1 to 10
a=1
while a<=10:
  print(a,end = ' ')
  a+=1 #a=a+1
```

```
# read the N as input
# Print the sum of N nartural numbers
# 5 -- 15
n=int(input(''))
a=1
sum=0
while a<=n:
  sum=sum+a #sum+=a
  a+=1
print(sum)
```

```
#read the input as N
# print only even numbers from 1 to N
# 10--2 4 6 8 10 
n=int(input(' '))
a=1
while a<=n:
  if a % 2 == 0:
    print(a,end=' ')
  a +=1
```

```
# read the input as N 
# print the factors or divisors of number N
# 6--1 2 3 6
# 12 -- 1 2 3 4 6 12
#24 -- 1 2 3 4 6 8 12 24
n=int(input(' '))
i=1
while i<=n:
  if n%i==0:
    print(i,end = '')
  i+=1
```
